# **PLANO XZ ----» $\phi=0$**

```python
# Parámetros de la simulación
plane = "XZ"
angle = "phi=0"
num_snapshots = 10
dir_path = "/tmp/p3disoj"


# Creamos la instancia de la simulación
simxz = Simulation(plane, angle, num_snapshots, dir_path)
```

## **Streamlines**

```python
# Crear la animación del campo de densidad
# Crear instancia de Visualize
visualizer = Visualize(simxz.data_handler)
visualizer.vel_streamlines(

    # Límites espaciales de la animación
    var1_min=0.5,
    var1_max=1.47,
    var2_min=0.006,
    var2_max=0.25,
    res=120,  # Resolución del mesh
    time=0.5  # Instante de tiempo para hallar el campo interpolado
)
```

```python
import numpy as np
# Crear la animación del campo de densidad
# Crear instancia de Visualize
visualizer = Visualize(simxz.data_handler)
visualizer.vel_streamlines_vid(
    var1_min=0.5,
    var1_max=1.47,
    var2_min=0.006,
    var2_max=0.25,
    res=120,
    time_array=np.linspace(0,1,10)
)
```

## **Simulación npi=12000 partículas**

```python
# Crear instancia de la simulación
sim =  Simulation(plane, angle, num_snapshots, dir_path)

# Ejecutar la simulación
sim.run_simulation(
    res=120,

    # Límites espaciales para la simulación
    var1_min=0.5,
    var1_max=1.47,
    var2_min=0.006,
    var2_max=0.25,
    ts=600,  # Número de particiones de tiempo
    npi=12000,
    max_lifetime=80,        # Tiempo de vida de las partículas (en frames)
    generation_interval=50  # Intervalo entre generación de nuevas partículas (en frames)
)
```

# **PLANO XZ ----» $\theta=1.56$ [rad]**

```python
# Parámetros de la simulación
plane = "XY"
angle = "theta=1.56"   # radianes
num_snapshots = 10
dir_path = "/tmp/p3disoj/"


# Creamos la instancia de la simulación
simxy = Simulation(plane, angle, num_snapshots, dir_path)
```
